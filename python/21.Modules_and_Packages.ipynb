# Importing Modules in Python: Modules and Packages
In Python, modules and packages help organize and reuse code.

```python
import math # Import a package or module , inside this we have many mathematical functions
math.sqrt(16)
```

```python
math.factorial(5)
```

```python
math.log2(16)
```

```python
from math import sqrt,pi,log2 # here from the package we are importing 3 functions ( because we only need these )
# Along with that we can call them directly
print(sqrt(16))
print(sqrt(25))
print(log2(12))
print(pi)
```
