# Faker is a Python package that allows us to generate your own fake dataset.By using this package we can generate random data such as Name,Address,Email,City,Country etc. 
This package supports multiple locations and languages to generate data based on locality.

Faker is not a standard python library,so first you need to install it using

```python
from faker import Faker
#Faker instance
fake=Faker()
print('Name:',fake.name())
print('Date of Birth:',fake.date_of_birth())
print('Address:',fake.address())
print('Country:',fake.country())
print('Email:',fake.email())
```

```python
## Data Serialization
import json # In order to work with json, we use json
data={'name':'Pritam','age':30} # my dict data

json_str=json.dumps(data) # converting the Dict to Json, it will give us a string
print(json_str)
print(type(json_str)) # Type of the json dump is string

parsed_data=json.loads(json_str) # Convert from string to Dict
print(parsed_data)
print(type(parsed_data))
```

# We have a csv file where e do not have any data


```python
## csv

import csv # To work with csv file

with open('example.csv',mode='w',newline='') as file: # Opening which file in which mode , w : write mode
    writer=csv.writer(file) # creating the writer
    writer.writerow(['name','age'])  # writing the First row
    writer.writerow(['Pritam',30]) # Second Row
    writer.writerow(['Swagata',28]) # Third Row
    writer.writerow(['Edhikaa',3]) # Fouth Row

with open('example.csv',mode='r') as file: # reading the file
    reader=csv.reader(file) # reader function
    for row in reader: # looping the rows and printing it
        print(row)
```

# Also if the .csv file is not there we can create it with the same code

```python
import csv # To work with csv file

with open('test.csv',mode='w',newline='') as file: # Creating a file named test.csv with write mode , w : write mode
    writer=csv.writer(file) # creating the writer
    writer.writerow(['name','age'])  # writing the First row
    writer.writerow(['Rahul',33]) # Second Row
    writer.writerow(['Swagata',28]) # Third Row
    writer.writerow(['Edhikaa',3]) # Fouth Row

with open('test.csv',mode='r') as file: # reading the file
    reader=csv.reader(file) # reader function
    for row in reader: # looping the rows and printing it
        print(row)
```
